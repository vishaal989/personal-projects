```python
cols_to_drop = ['investor_permalink', 'company_permalink', 'investor_category_code','funded_month','funded_quarter', 'funded_year']
cols_to_include = all_df.columns.drop(cols_to_drop)
```

```python
cols_to_include
```

```python
unique_val_prct = {}
for col in cols_to_include:
    pct = len(all_df[col].value_counts())/len(all_df[col])
    if pct <=0.15:
        unique_val_prct[col] = pct
```

```python
unique_val_prct
```

```python
cols_to_cat = list(unique_val_prct.keys())
```

```python
cols_to_cat
```

```python
cols_to_cat
cols_to_cat.remove('funded_at')
cols_to_cat.remove('raised_amount_usd')
```

```python
memory_optimized = 0
all_data_op = []
df_chunk_op = pd.read_csv("crunchbase-investments.csv",chunksize=5000,encoding='ISO-8859-1',usecols=cols_to_include)

for chunk in df_chunk_op:
    for col in chunk.columns:
        if col in cols_to_cat:
            chunk[col] = chunk[col].astype("category")
        elif col == 'raised_amount_usd':
            chunk[col] = pd.to_numeric(chunk[col],downcast='float')
        elif col == 'funded_at':
            chunk[col] = pd.to_datetime(chunk[col])
        
    all_data_op.append(chunk)
    mem = chunk.memory_usage(deep=True).sum()/(1024*1024)
    #print(mem)
    memory_optimized += mem

mem_before_op = all_df.memory_usage(deep=True).sum()/(1024*1024)
print("\nMemory utilised by DF BEFORE optimization: " + str(round(mem_before_op,1)) + " MB")
print("\nMemory utilised by DF AFTER optimization: " + str(round(memory_optimized,1)) + " MB")

```

```python
import sqlite3
conn = sqlite3.connect('crunchbase.db')
chunk_iter = pd.read_csv('crunchbase-investments.csv', chunksize=5000, encoding='ISO-8859-1',usecols=cols_to_include)

for chunk in chunk_iter:
    chunk.to_sql("investments", conn, if_exists='append', index=False)
```

```python
q = '''
    select company_name,
            company_category_code,
            investor_name,
            funding_round_type,
            raised_amount_usd   
    from investments;
    '''
popular_category = []
top_investors = []
funding_rounds = []

chunk_iter = pd.read_sql(q, conn, chunksize=250)
for chunk in chunk_iter:
    #popular category
    pop_cat = chunk[['company_category_code','raised_amount_usd']].groupby(['company_category_code']).sum()
    popular_category.append(pop_cat)
    #top investor
    top_investor = chunk[['investor_name','raised_amount_usd']].groupby(['investor_name']).sum()
    top_investors.append(top_investor)
    #popular funding round
    funding_rnd = chunk[['funding_round_type','raised_amount_usd']].groupby(['funding_round_type']).sum()
    funding_rounds.append(funding_rnd)
    
    
    
    
    
```

```python
category_df = pd.concat(popular_category)
```

```python
category_df = category_df.reset_index()
```

```python
category_df = category_df.groupby(['company_category_code']).sum().sort_values(by=['raised_amount_usd'],ascending=False)
```

```python
category_df
category_df.style.format("{:.1f}")
```

```python
investor_df = pd.concat(top_investors)
investor_df = investor_df.reset_index()
investor_df = investor_df.groupby(['investor_name']).sum().sort_values(by=['raised_amount_usd'],ascending=False)
investor_df.style.format("{:.1f}")
```

```python
funding_round_df = pd.concat(funding_rounds)
funding_round_df = funding_round_df.reset_index()
funding_round_df = funding_round_df.groupby(['funding_round_type']).sum().sort_values(by=['raised_amount_usd'],ascending=False)
funding_round_df.style.format("{:.1f}")
```
